```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("greenhouse.csv")
df.country_or_area.unique()
```

```python
df.head()
Australia = df[df.country_or_area == "Australia"]
Ukraine = df[df.country_or_area == "Ukraine"]
```

```python
australia = Australia.groupby(["year"])["value"].mean()
australia = pd.DataFrame(australia, index=None)
australia.rename({"value": "mean_value"}, axis=1, inplace=True)
australia
```

```python
australia["year"] = australia.index
```

```python
australia
```

```python
fig, ax = plt.subplots()
ax.plot(australia.year, australia.mean_value)
ax.set_xlabel("year")
ax.set_ylabel("mean_value")
import streamlit as st
with st.container():
    st.pyplot(fig)
```
